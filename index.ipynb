```python
import pandas as pd
```

```python
surveys_df = pd.read_csv("surveys.csv")
```

```python
species_counts = surveys_df.groupby('species_id')['record_id'].count()
```

```python
# make sure figures appear inline in Ipython Notebook
%matplotlib inline
# create a quick bar chart
species_counts.plot(kind='bar');
```
